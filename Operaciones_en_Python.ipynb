# Operaciones aritmÃ©ticas bÃ¡sicas

```python
#Operadores aritmeticos en Python
x = 10; y = 3
print("Operadores aritmÃ©ticos")
print("x+y =", x+y)   #Suma, resultado: 13
print("x-y =", x-y)   #Resta, resultado: 7
print("x*y =", x*y)   #Multiplicacion, resultado: 30
print("x/y =", x/y)   #DivisiÃ³n, resultado: 3.3333333333333335
print("x%y =", x%y)   #MÃ³dulo, resultado: 1
print("x**y =", x**y) #Exponente, resultado: 1000
print("x//y =", x//y) #Cociente, resultado: 3
```

# Orden de aplicaciÃ³n

En los ejemplos anteriores simplemente hemos aplicado un operador a dos nÃºmeros sin mezclarlos entre ellos. TambiÃ©n es posible tener varios operadores en la misma lÃ­nea de cÃ³digo, y en este caso es muy importante tener en cuenta las prioridades de cada operador y cual se aplica primero. Ante la duda siempre podemos usar parÃ©ntesis, ya que todo lo que estÃ¡ dentro de un parÃ©ntesis se evaluarÃ¡ conjuntamente, pero es importante saber las prioridades.

El orden de prioridad serÃ­a el siguiente para los operadores aritmÃ©ticos, siendo el primero el de mayor prioridad:

    () ParÃ©ntesis
    ** Exponente
    -x NegaciÃ³n
    * / // MultiplicaciÃ³n, DivisiÃ³n, Cociente, MÃ³dulo
    + - Suma, Resta

```python
def operacion_sencilla(numero1, numero2, operacion):
    """
    Esta funciÃ³n toma dos nÃºmeros y una operaciÃ³n aritmÃ©tica y devuelve el resultado
    de la operaciÃ³n. La operaciÃ³n se especifica como una cadena, que debe ser uno de
    los siguientes valores: '+', '-', '*', '/'.
    """
    if operacion == '+':
        resultado = numero1 + numero2
    elif operacion == '-':
        resultado = numero1 - numero2
    elif operacion == '*':
        resultado = numero1 * numero2
    elif operacion == '/':
        resultado = numero1 / numero2
    else:
        resultado = None

    return resultado

#Resolver la operaciÃ³n 4 + 3
resultado = operacion_sencilla(4, 3, '+')
print(resultado)  # Salida: 7

# Resolver la operaciÃ³n 9 - 6
resultado = operacion_sencilla(9, 6, '-')
print(resultado)  # Salida: 3

# Resolver la operaciÃ³n 5 * 2
resultado = operacion_sencilla(5, 2, '*')
print(resultado)  # Salida: 10

# Resolver la operaciÃ³n 12 / 4
resultado = operacion_sencilla(12, 4, '/')
print(resultado)  # Salida: 3.0
```
