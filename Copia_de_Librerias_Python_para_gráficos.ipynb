```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
datos = np.random.normal(loc=0, scale=1, size=1000)

# Gráfico de histograma
plt.hist(datos, bins='auto', alpha=0.7, rwidth=0.85)
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Frecuencia')
plt.grid(True)
plt.show()

# Gráfico de caja y bigote (boxplot)
plt.boxplot(datos)
plt.ylabel('Valor')
plt.title('Diagrama de Caja y Bigote')
plt.grid(True)
plt.show()

# Gráfico de diagrama de dispersión (scatter plot)
x = np.random.normal(loc=0, scale=1, size=100)
y = np.random.normal(loc=0, scale=1, size=100)
plt.scatter(x, y)
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.title('Diagrama de Dispersión')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo: Tiempos de respuesta del servidor (en milisegundos)
tiempos_respuesta = np.random.normal(loc=100, scale=20, size=1000)

# Gráfico de histograma
plt.hist(tiempos_respuesta, bins='auto', alpha=0.7, rwidth=0.85)
plt.xlabel('Tiempo de Respuesta (ms)')
plt.ylabel('Frecuencia')
plt.title('Distribución de Tiempos de Respuesta del Servidor')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo: Latencia de tres servidores (en milisegundos)
servidor_1 = np.random.normal(loc=100, scale=15, size=100)
servidor_2 = np.random.normal(loc=120, scale=20, size=100)
servidor_3 = np.random.normal(loc=80, scale=10, size=100)

# Gráfico de diagrama de caja y bigote
plt.boxplot([servidor_1, servidor_2, servidor_3], labels=['Servidor 1', 'Servidor 2', 'Servidor 3'])
plt.ylabel('Latencia (ms)')
plt.title('Comparación de Latencia en Servidores')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo: Uso de CPU y latencia del servidor
uso_cpu = np.random.normal(loc=50, scale=10, size=100)  # Uso de CPU en porcentaje
latencia = np.random.normal(loc=100, scale=20, size=100)  # Latencia en milisegundos

# Gráfico de diagrama de dispersión
plt.scatter(uso_cpu, latencia)
plt.xlabel('Uso de CPU (%)')
plt.ylabel('Latencia (ms)')
plt.title('Relación entre Uso de CPU y Latencia del Servidor')
plt.grid(True)
plt.show()
```

```python
#Este gráfico puede representar el uso de ancho de banda (en MB) en diferentes intervalos de tiempo durante un día típico.
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo: Uso de ancho de banda (en MB) en diferentes horas
horas = np.arange(0, 24, 1)  # Horas del día
ancho_banda = np.random.normal(loc=100, scale=20, size=24)  # Uso de ancho de banda en MB

# Gráfico de líneas
plt.plot(horas, ancho_banda, marker='o')
plt.xlabel('Hora del Día')
plt.ylabel('Ancho de Banda (MB)')
plt.title('Monitoreo del Uso de Ancho de Banda en la Red')
plt.grid(True)
plt.show()
```

```python
#Gráfico de barras
# Este gráfico compara el uso promedio de recursos (CPU y Memoria) en diferentes aplicaciones del sistema.

import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo: Uso promedio de CPU y memoria para diferentes aplicaciones
aplicaciones = ['App1', 'App2', 'App3', 'App4']
uso_cpu = [35, 50, 65, 45]  # Uso de CPU en porcentaje
uso_memoria = [400, 600, 800, 550]  # Uso de memoria en MB

# Gráfico de barras
x = np.arange(len(aplicaciones))  # Etiquetas de los ejes X
width = 0.35  # Ancho de las barras

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, uso_cpu, width, label='CPU (%)')
rects2 = ax.bar(x + width/2, uso_memoria, width, label='Memoria (MB)')

# Añadir etiquetas y título
ax.set_xlabel('Aplicaciones')
ax.set_ylabel('Uso de Recursos')
ax.set_title('Comparación del Uso de CPU y Memoria en Diferentes Aplicaciones')
ax.set_xticks(x)
ax.set_xticklabels(aplicaciones)
ax.legend()

plt.grid(True)
plt.show()
```

```python
#Gráfico circular (Pie chart)
#Este gráfico muestra la proporción de diferentes tipos de incidentes reportados al soporte técnico durante un mes.

import matplotlib.pyplot as plt

# Datos de ejemplo: Proporción de tipos de incidentes reportados
tipos_incidentes = ['Errores de red', 'Problemas de hardware', 'Problemas de software', 'Consultas de usuario']
porcentajes = [30, 25, 35, 10]

# Gráfico circular
plt.pie(porcentajes, labels=tipos_incidentes, autopct='%1.1f%%', startangle=90)
plt.title('Distribución de Tipos de Incidentes en el Soporte Técnico')
plt.show()
```

```python
#Gráfico de área
# Este gráfico muestra la evolución del tráfico en un sitio web, medido en visitas, a lo largo de una semana

import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo: Número de visitas diarias a lo largo de una semana
dias = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
visitas = np.random.normal(loc=1000, scale=200, size=7)

# Gráfico de área
plt.fill_between(dias, visitas, color="skyblue", alpha=0.4)
plt.plot(dias, visitas, color="Slateblue", alpha=0.6, linewidth=2)

plt.title('Evolución del Tráfico Web en una Semana')
plt.xlabel('Día')
plt.ylabel('Número de Visitas')
plt.grid(True)
plt.show()
```

```python
#Heatmap (mapa de calor)
#Este gráfico muestra un mapa de calor que representa el uso de CPU en diferentes servidores durante una semana.

import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo: Uso de CPU en porcentaje para 5 servidores durante 7 días
servidores = ['Servidor 1', 'Servidor 2', 'Servidor 3', 'Servidor 4', 'Servidor 5']
dias = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
uso_cpu = np.random.randint(20, 100, size=(5, 7))

# Gráfico de mapa de calor (heatmap)
plt.imshow(uso_cpu, cmap='coolwarm', aspect='auto')

# Etiquetas de los ejes
plt.xticks(np.arange(len(dias)), dias)
plt.yticks(np.arange(len(servidores)), servidores)

# Título y barra de color
plt.title('Uso de CPU en Diferentes Servidores durante la Semana')
plt.colorbar(label='Uso de CPU (%)')
plt.show()
```
